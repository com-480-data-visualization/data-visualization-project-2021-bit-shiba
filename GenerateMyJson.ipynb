```python
import numpy as np
import pandas as pd
import json
```

```python
# import dataset and preprocess our data
df = pd.read_csv(r"content/pokemon_gen_1_to_8.csv")
df.drop(columns=['Unnamed: 0','german_name','japanese_name'],inplace=True)
df = df.fillna('/')
```

```python
interested = ['hp','attack','defense','sp_attack','sp_defense','speed']
for c in interested:
    df[c] = df[c].astype(float)
df = df.assign(species_strength = df[interested].sum(axis=1)) 
```

```python
df.head()
```

```python
violindata = df[['type_1','species_strength']]
violindata.columns = ['Type', 'SpeciesStrength']
result = violindata.to_json(orient="records")
parsed = json.loads(result)
with open('content/violindata.txt', 'w') as outfile:
    json.dump(parsed, outfile, indent=4)
```
